So I'm going to be making a new dataframe(a copy of the original one), where I'm free to make changes without messing with the original dataframe.

So what do I want to filter from this data to be able to analyze? I want to be able to return only specific information of a value in a column. I want to filter the data in which the State has the value 'New York' and the Subgroup has the values is '18-29 years'. 

So i'll be making a variable in order to filter the set. I want to go ahead and put this new filtered set into a seperate dataframe, which I'll call 'df2'. 

```python
df2 = df1[(df1.Subgroup == '18 - 29 years') | (df1.State == 'New York')]
df2
#Similar code df2 = df1.loc[(df1.Subgroup == '18 - 29 years') | (df1.State == 'New York')]
# group = ['18 - 29 years', 'New York'] 
# df2[df2['Subgroup'].isin(group)]
```

In order to filter two values, you need to create a list. So I'm going to create a variable that contains the lists of filtered values, then use " .isin() " to apply it to my filtering code. Pandas isin() is used to filter data frames and selects the rows that have particular values in a given column.

```python
group = ['18 - 29 years', 'New York'] 
df2[df2['Subgroup'].isin(group)]
```
